```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.datasets import load_wine
import seaborn as sns 
import matplotlib.pyplot as plt

data = load_wine(return_X_y = True)
standard_scaler = preprocessing.StandardScaler()
standard = standard_scaler.fit_transform(data[0])

df = pd.DataFrame(standard)
df_cov = df.cov()
df_cov
```

```python
df_cov.style.background_gradient(cmap='rainbow', axis=None)
```

```python
f, ax = plt.subplots(figsize=(10, 8))   
sns.heatmap(df_cov, mask=np.zeros_like(df.corr(), dtype=np.bool),                   
            cmap=sns.diverging_palette(220, 10, as_cmap=True), square=True, ax=ax)


```
