# Load libraries and dataset

```python
import networkx as nx
import pandas as pd

bitcoinData = pd.read_csv('downloads/soc-sign-bitcoinotc.csv')
```

```python
bitcoinData
```

# Data preprocessing

```python
bitcoinList = []
bitcoinList.append((6,2,4))

for i in range(len(bitcoinData['6'])):
    bitcoinList.append((bitcoinData['6'][i],bitcoinData['2'][i],bitcoinData['4'][i]))  
```

```python
bitcoinList
```

```python
G = nx.DiGraph()
G.add_weighted_edges_from(bitcoinList)
```

# 6-c. The two nodes with the highest PageRank

```python
nx.pagerank(G, alpha = 0.3, max_iter = 1000)
```

```python
pr = nx.pagerank(G, alpha = 0.3, max_iter = 1000)
sort_pr = sorted(pr.items(), key=lambda x: x[1], reverse=True)
```

```python
print(sort_pr[0])
print(sort_pr[1])
```

```python
# The two nodes with the highest PageRank: node 35 and node 2642
# Same as the two nodes which have the highest Betweeness centrality
# The alpha I used is kind of low but if I didn't use 0.3 it will take forever for it to converge.
```

# 9. Extra credit question

```python
# For this extra credit question I created the graph again because to use the louvain algorithm we need
# undirected gragh and unweighted edges.
# reference: https://python-louvain.readthedocs.io/en/latest/
```

```python
nbitcoinList = []
nbitcoinList.append((6,2))

for i in range(len(bitcoinData['6'])):
    nbitcoinList.append((bitcoinData['6'][i],bitcoinData['2'][i]))  
```

```python
nbitcoinList
```

```python
nG = nx.Graph()
nG.add_edges_from(nbitcoinList)
```

```python
from community import community_louvain
import matplotlib.cm as cm
import matplotlib.pyplot as plt

# compute the best partition
partition = community_louvain.best_partition(nG)

# draw the graph
pos = nx.spring_layout(G)

# color the nodes according to their partition
cmap = cm.get_cmap('viridis', max(partition.values()) + 1)
nx.draw_networkx_nodes(G, pos, partition.keys(), node_size=40,
                       cmap=cmap, node_color=list(partition.values()))
nx.draw_networkx_edges(G, pos, alpha=0.5)
plt.show()
```

```python
# The graph above are the communities found using the Louvain algorithm.
```
