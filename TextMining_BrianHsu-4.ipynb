```python
# Extracting the values

import re

text = 'I am 20 years old. My previous license plate number was 4XUI302 and my new one is 3ABC278. My ID is J987492 and my address is 123 Main street, San Jose, CA. Please email me at myemail123+spam@google.com'

license_plate_nos = re.findall(r"[\d][A-Z]{3}[\d]{3}",text)
print("Old license plate number: " + license_plate_nos[0])
print("New license plate number: " + license_plate_nos[1])

ID = re.findall(r"[A-Z][\d]{6}",text)
print("ID: " + ID[0])

mailing_address = re.findall(r"[\s][\d]{3}[\s\w,][\s\w,]+[A-Z]{2}",text)
print("Mailing Address: " + mailing_address[0])

email = re.findall(r"[\w.+-]+@[\w.+-]+",text)
print("Email: " + email[0])

# abbreviating license plate number to LP_NUM

text = text.replace("license plate number","LP_NUM")
text



```

```python
# removing punctuation

import nltk
import string
from nltk.corpus import movie_reviews
nltk.download('movie_reviews')
s = movie_reviews.words()
ns = []
for w in s:
    if w not in string.punctuation:
        ns.append(w)
    else:
        continue
ns 
```

```python
# number of unique words

unique_ns = set(ns)
len(unique_ns)
```

```python
# top 20 most frequently encountered words in the corpus

fdist = nltk.FreqDist(ns)
fdist.most_common(20)
```

```python
#removing stop words
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = stopwords.words('english')
ns_nsw = []
for w in ns:
    if w not in stop_words:
        ns_nsw.append(w)
    else:
        continue
ns_nsw

```

```python
# number of unique words after removing stop words

unique_ns_nsw = set(ns_nsw)
len(unique_ns_nsw)
```

```python
# top 20 most frequently encountered words in the corpus after removing the stop words

fdist = nltk.FreqDist(ns_nsw)
fdist.most_common(20)
```

```python
# words that are used only once in the corpus (show the first 20 words returned by the hapaxes() function)

once = fdist.hapaxes()
once[0:20]
```

```python
# number of unique words after stemming is applied

from nltk.stem import PorterStemmer 

ps = PorterStemmer()
stemmed = []
for w in ns_nsw:
    s = ps.stem(w)
    stemmed.append(s)
stemmed_set = set(stemmed)
len(stemmed_set)
```

```python
# number of unique words after lemmatization is applied

from nltk.stem import WordNetLemmatizer 

nltk.download('wordnet')
lemmatizer = WordNetLemmatizer() 
lemmatized = []
for w in ns_nsw:
    s = lemmatizer.lemmatize(w)
    lemmatized.append(s)
lemmatized_set = set(lemmatized)
len(lemmatized_set)
```
